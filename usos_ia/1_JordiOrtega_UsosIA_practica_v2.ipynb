```python
# La tarea seleccionada es "Responder correos electrónicos frecuentes de clientes en una empresa de telecomunicaciones que se quejan de un mal servicio de internet".
# Esta tarea es repetitiva y puede automatizarse para ahorrar tiempo y mejorar la comunicación y atención al cliente.
```

```python
################################################################################################################################
############################## REDACCIÓN PROMPT ################################################################################
################################################################################################################################
```

```python
############################# TÉCNICAS PROMPTING ###############################################################################

# Zero-Shot -> Implica generar una respuesta sin proporcionar ejemplos o contexto
#     Retrieval -> Trata a los modelos de IA como motores de búsqueda, formulando preguntas específicas para obtener respuestas detalladas
# One-Shot -> Consiste en proporcionar un solo ejemplo o contexto para guiar la respuesta del modelo.
#             Es efectivo para tareas específicas donde un ejemplo puede servir como guía clara
# CoT (Chain of Thought) ->  Para resolver problemas complejos o tareas de razonamiento se fomenta la división de estos en partes más pequeñas y manejables
#     Prompt Activo -> Explica el proceso intermedio que llevó a esa conclusión
# Auto-consistencia -> Se basa en generar varias cadenas de pensamiento para una tarea, luego seleccionar la respuesta más consistente entre ellas
# Prompt de conocimiento -> Es la capacidad de los modelos de lenguaje (LM) para generar conocimiento que ayude a hacer predicciones precisas
# ToT (Tree of Thoughts) -> Sistema para estructurar el razonamiento de la IA de forma secuencial y detallada para abordar problemas complejos
# RAG (Generación de Recuperación Aumentada) -> Combina un modelo de recuperación de información con un generador de texto. Este método permite a los modelos acceder a origenes externos
# RolePlay (Expert prompting) -> Indicarle a una IA que asuma un rol para que actúe como un experto de una materia
# ReACT -> Combina el razonamiento (pensar) y las acciones específicas (actuar) para resolver tareas complejas
```

```python
############################# ESTRUCTURA PROMPT ###############################################################################

# ------> Instrucción clara
# Eres un asistente de atención al cliente, sin acceso a bases de datos de clientes, del sector telecomunicaciones especializado en tratar incidencias técnicas.
# Tu misión es redactar correos electrónicos claros, amables y profesionales en respuesta a las consultas más comunes de los clientes.

# ------> Contexto
# El cliente está enfadado debido a que no tiene disponible el servicio de internet contratado y está pidiendo que le restauren la conexión a internet.

# ------> Formato de salida
# Proporciona una respuesta de máximo 200 palabras, con un tono profesional y accesible.
# Divide el texto en párrafos claros para facilitar la lectura.
# Lista las acciones recomendadas para resolver la incidencia.
# Incluye un cierre amigable que invite al cliente a contactarnos si necesita más ayuda.

# ------> Ejemplo
# Consulta del cliente:
# Hola, Llevo 15 minutos sin servicio de red y esto hace que no pueda trabajar.
# Necesito que me indiquen cómo resolver la incidencia desde mi casa o si se trata de un problema masivo que digan cuando va a estar
# restaurado el servicio. Mi número de cliente es el 1191283120.
```

```python
############################# RESPUESTA ESPERADA ###############################################################################

# La practica pretende dar respuesta a 2 niveles de exigencia: respuesta genérica y respuesta personalizada.

###### 1. respuesta genérica ######
# Rápida y más fácil de automatizar pero puede generar frustración si el cliente ya intentó aplicar estas soluciones o si su problema es parte de una incidencia masiva que no puede resolver por sí mismo.

# Estimado/a cliente,

# Gracias por contactarnos. Lamentamos los inconvenientes causados por la interrupción de su servicio de internet. Por favor, siga estos pasos para intentar resolver el problema desde casa:

# Reinicie su módem apagándolo y encendiéndolo después de 30 segundos.
# Verifique que los cables estén correctamente conectados.
# Intente conectarse con otro dispositivo para descartar problemas con su equipo.

# En caso de que ninguna de estas acciones solucione el problema, contáctenos nuevamente para continuar con la resolución de la incidencia.

# Atentamente,
# Equipo de soporte técnico

###### 2. respuesta personablizada ######
# Muestra atención adaptada al usuario en cuestión pero requiere de acceso a los sistemas internos de la compañía

# Estimado/a [Nombre del cliente],

# Lamentamos los inconvenientes con su servicio de internet. Hemos verificado su número de cliente ([xxxxxxxxxxx]) y constatado que en su zona actualmente no hay una incidencia masiva que afecta la conexión.

# Le sugerimos siga estos pasos para intentar resolver el problema desde casa:

# Reinicie su módem apagándolo y encendiéndolo después de 30 segundos.
# Verifique que los cables estén correctamente conectados.
# Intente conectarse con otro dispositivo para descartar problemas con su equipo.

# Si el problema persiste, no dude en contactarnos nuevamente. Estamos aquí para ayudarle.

# Atentamente,
# [Nombre del agente o equipo]
```

```python
############################# TÉCNICAS PROMPTING USADAS ########################################################################

# Podemos identificar 2 técnicas de prompt si el objetivo es una respuesta genérica:
# 1. RolePlay
# Podemos identificar la estructura tipo: contexto (asistente de un sector determinado con una especialidad), instrucción (redactar correos electrónicos), input (responde a las preguntas de problemas técnicos de los clientes)
# y output (formato/estructura del texto).

# 2. Zero-Shot
# Aunque se proporciona un contexto, el prompt no incluye un ejemplo de tarea resuelta.

# Si el objetivo fuera tener una respuesta personalizada, el prompt debería de incluir una tercera técnica de prompting:
# 3. RAG
```
